### Using TFIDF

```python
#converting our text columns to tfidf values
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer()
X_train_tfidf = tfidf_vect.fit_transform(people.text)
```

```python
X_train_tfidf
```

```python
tfidf_vect.get_feature_names()
```

```python
X_train_tfidf
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn = NearestNeighbors(n_neighbors=10,p=2)
```

```python
nn_fitted = nn.fit(X_train_tfidf)
```

```python
nn_fitted
```

## minkowski distance

- metric = 'minkowski
- p = 2 (it can be change) "2 is deafult value"

- manhattan distance
- eculidian distance

```python
# Manhattan distance
x = [1,0,0]
y = [0,1,0]

distance = 0

for i,j in zip(x,y):
    distance += abs(i-j) ** 1
    
print(distance ** (1/1))
```

```python
#eculidian distance
x = [1,0,0]
y = [0,1,0]

distance = 0

for i,j in zip(x,y):
    distance += (i-j) ** 2
    
print(distance ** (1/2))
```

```python
#minkowski - function - generic function
x = [1,0,0]
y = [0,1,0]
p=2
distance = 0

for i,j in zip(x,y):
    distance += (i-j) ** p
    
print(distance ** (1/p))
```
